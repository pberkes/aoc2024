```python
%matplotlib inline

import itertools
import re

import matplotlib.pyplot as plt
import networkx as nx
import numpy as np


def map_array(map_txt):
    rows = map_txt.splitlines()
    map_ = np.array([list(r) for r in rows])
    return map_


def print_map(map_):
    for row in map_:
        print(''.join(row))

        
def out_of_bounds(pos, n, m):
    return (pos[0] < 0) | (pos[0] >= n) | (pos[1] < 0) | (pos[1] >= m)


directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]
```

```python
txt = """
OOOOO
OXOXO
OOOOO
OXOXO
OOOOO
""".strip()
```

```python
txt = """
RRRRIICCFF
RRRRIICCCF
VVRRRCCFFF
VVRCCCJFFF
VVVVCJJCFE
VVIVCCJJEE
VVIIICJJEE
MIIIIIJJEE
MIIISIJEEE
MMMISSJEEE
""".strip()
```

```python
with open('input.txt', 'r') as f:
    txt = f.read().strip()
```

# Part 1

```python
map_ = map_array(txt)
plants = np.unique(map_)
n, m = map_.shape
```

```python
not_visited = set(itertools.product(range(n), range(m)))

perimeters = []
areas = []
while len(not_visited) > 0:
    pos = not_visited.pop()

    plant = map_[*pos]
    perimeter = 0
    area = 0
    visited = set()

    consider_queue = set()
    while True:
        visited.add(pos)
        area += 1
        for d in directions:
            next_pos = pos[0] + d[0], pos[1] + d[1]
            if out_of_bounds(next_pos, n, m):
                perimeter += 1
            elif map_[*next_pos] == plant:
                if not next_pos in visited:
                    consider_queue.add(next_pos)
            else:
                perimeter += 1
        if len(consider_queue) == 0:
            break
        pos = consider_queue.pop()
        not_visited.remove(pos)
        
    perimeters.append(perimeter)
    areas.append(area)
```

```python
sum(areas[idx] * perimeters[idx] for idx in range(len(areas)))
```

```python
txt = """
AAAA
BBCD
BBCC
EEEC
""".strip()
```

```python
txt = """
AAAAAA
AAABBA
AAABBA
ABBAAA
ABBAAA
AAAAAA
""".strip()
```

```python
with open('input.txt', 'r') as f:
    txt = f.read().strip()
```

```python
map_ = map_array(txt)
plants = np.unique(map_)
n, m = map_.shape
```

```python
not_visited = set(itertools.product(range(n), range(m)))

edges = []
areas = []
while len(not_visited) > 0:
    pos = not_visited.pop()

    plant = map_[*pos]
    area = 0
    visited = set()
    outside = []

    consider_queue = set()
    while True:
        visited.add(pos)
        area += 1
        for d in directions:
            next_pos = pos[0] + d[0], pos[1] + d[1]
            if out_of_bounds(next_pos, n, m):
                outside.append((d[0], d[1], next_pos[0], next_pos[1]))
            elif map_[*next_pos] == plant:
                if not next_pos in visited:
                    consider_queue.add(next_pos)
            else:
                outside.append((d[0], d[1], next_pos[0], next_pos[1]))
        if len(consider_queue) == 0:
            break
        pos = consider_queue.pop()
        not_visited.remove(pos)
        
        
    g = nx.Graph()
    g.add_nodes_from(outside)
    for node1, node2 in itertools.combinations(outside, 2):
        dist = sum([abs(node1[i] - node2[i]) for i in range(len(node1))])
        if dist == 1:
            g.add_edge(node1, node2)
            
    nedges = nx.number_connected_components(g)
    edges.append(nedges)
    areas.append(area)
```

```python
areas, edges
```

```python
sum(areas[idx] * edges[idx] for idx in range(len(areas)))
```
